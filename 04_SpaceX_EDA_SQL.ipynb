# Connect to the database using SQLite database

```python
import sqlite3
import pandas as pd 
```

```python
#create a SQLite database with name data
db = sqlite3.connect('data.db')
```

```python
# some funcrtion to deal with database 

def read(db, read_):
    print('Read')
    cursor = db.cursor()
    cursor.execute(read_)
    for row in cursor:
        print(f'row = {row}')
    print()
    
# function to create in postgre database     
def create(db, create_):
    cursor = db.cursor() # create cursor object
    cursor.execute(create_) # execute query
    db.commit() # commit query to database
    print('Table have been created successfull!!!')
    #read(db)
    
# function to insert in postgre database     
def insert(db, insert_):
    cursor = db.cursor()
    cursor.execute(insert_)
    db.commit()
    print('Records have been successfully inserted!!!')
    #read(db)
    
# function to update table
def update(db, update_):
    print('Update')
    cursor = db.cursor()
    cursor.execute(update_)
    db.commit()
    #read(db)
    
# function to delete in postgre database
def delete(db, delete_):
    print('Delete')
    cursor = db.cursor()
    cursor.execute(delete_)
    db.commit()
    #read(db)

# close the cursor and dbection to the server 
def close():
    cursor.close()
    db.close()   
# function to create pandas dataframe
def create_pandas_df(sql_query, database=db):
    table = pd.read_sql_query(sql_query, database)
    return table
```

```python
# create teble spacex
create_ = '''
           
            CREATE TABLE if not exists SpaceX
                (
                    Date DATE NULL,
                    Time TIME NULL,
                    BoosterVersion VARCHAR(50) NULL,
                    LaunchSite VARCHAR(50) NULL,
                    Payload VARCHAR(100) NULL,
                    PayloadMassKG INT NULL,
                    Orbit VARCHAR(50) NULL,
                    Customer VARCHAR(100) NULL,
                    MissionOutcome VARCHAR(50) NULL,
                    LandingOutcome VARCHAR(100) NULL
                );
            '''
create(db, create_)
```

```python
# insert the value from csv file to database 

import csv

cursor = db.cursor()
with open('Spacex.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader) # Skip the header row.
    for row in reader:
        print(row)
        cursor.execute(
        f"INSERT INTO SpaceX VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)" , row
    )
db.commit()

print('CSV file inserted into database successfully!!!')
```

```python
# the result will be a dataFrame  
read_ = '''
        SELECT *
        FROM   SpaceX
    '''
spacex = create_pandas_df(read_, database=db)
spacex.head(10)
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
For Example "Landing_Outcome"**

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
sql_1 = '''
        SELECT DISTINCT LaunchSite 
        FROM SpaceX
'''
create_pandas_df(sql_1, database=db)
```

# Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
sql_2 = '''
        SELECT *
        FROM SpaceX
        WHERE LaunchSite LIKE 'CCA%'
        LIMIT 5
        '''
create_pandas_df(sql_2, database=db)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
sql_3 = '''
        SELECT SUM(PayloadMassKG) AS Total_PayloadMass
        FROM SpaceX
        WHERE Customer LIKE 'NASA (CRS)'
        '''
create_pandas_df(sql_3, database=db)
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
sql_4 = '''
        SELECT AVG(PayloadMassKG) AS Avg_PayloadMass
        FROM SpaceX
        WHERE BoosterVersion = 'F9 v1.1'
        '''
create_pandas_df(sql_4, database=db)
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
# %sql SELECT MIN(DATE) \
# FROM SPACEXTBL \
# WHERE LANDING__OUTCOME = 'Success (ground pad)'
sql_5 = '''
        SELECT MIN(Date) 
        FROM SpaceX
        WHERE LandingOutcome = 'Success (ground pad)'
        '''
create_pandas_df(sql_5, database=db)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
sql_6 = '''
        SELECT BoosterVersion
        FROM SpaceX
        WHERE LandingOutcome = 'Success (drone ship)'
            AND PayloadMassKG > 4000 
            AND PayloadMassKG < 6000
        '''
create_pandas_df(sql_6, database=db)
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
sql_7_1 = '''
        SELECT COUNT(MissionOutcome) 
        FROM SpaceX
        WHERE MissionOutcome LIKE 'Success%'
        '''

sql_7_2 = '''
        SELECT COUNT(MissionOutcome) 
        FROM SpaceX
        WHERE MissionOutcome LIKE 'Failure%'
        '''

display(create_pandas_df(sql_7_1, database=db))

create_pandas_df(sql_7_2, database=db)
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
sql_8 = '''
        SELECT BoosterVersion, PayloadMassKG
        FROM SpaceX
        WHERE PayloadMassKG = (
                                SELECT MAX(PayloadMassKG)
                                FROM SpaceX
                                )
        ORDER BY BoosterVersion
        '''
create_pandas_df(sql_8, database=db)
```

### Task 9

##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
# %sql SELECT substr(Date,4,2) as month, DATE,BOOSTER_VERSION, LAUNCH_SITE, [Landing _Outcome] \
# FROM SPACEXTBL \
# where [Landing _Outcome] = 'Failure (drone ship)' and substr(Date,7,4)='2015';
task_9 = '''
        SELECT BoosterVersion, LaunchSite, LandingOutcome
        FROM SpaceX
        WHERE LandingOutcome LIKE 'Failure (drone ship)'
            AND Date BETWEEN '2015-01-01' AND '2015-12-31'
        '''
create_pandas_df(task_9, database=db)
```

### Task 10

##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
%sql SELECT [Landing _Outcome], count(*) as count_outcomes \
FROM SPACEXTBL \
WHERE DATE between '04-06-2010' and '20-03-2017' group by [Landing _Outcome] order by count_outcomes DESC;
```
